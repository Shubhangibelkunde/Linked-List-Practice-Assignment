```python
1) Define a doubly linked list??
```

```python
- The DoublyLinkedList class maintains references to the head and tail of the list. 
- Additional methods would be implemented in the DoublyLinkedList class to perform various operations on the list.

- A doubly linked list is a data structure consisting of a sequence of elements, called nodes, where each node contains
two pointers/references: one pointing to the previous node and another pointing to the next node in the sequence. 
Unlike a singly linked list where each node only has a pointer to the next node, a doubly linked list allows traversal
in both forward and backward directions.

Com[onents of the doubly linked list = 
1)Node: Each element in the list is represented by a node.
    Each node contains data and two pointers: one to the previous node (prev pointer) and one to the next node (next pointer).

2)Head Pointer: A reference to the first node in the list.

3)Tail Pointer: A reference to the last node in the list.

4)Operations:
Insertion: Nodes can be inserted at the beginning, end, or at any position within the list.
Deletion: Nodes can be removed from any position within the list.
Traversal: Allows traversing the list in both forward and backward directions.
Searching: Searching for a specific element within the list.
```

```python
2) Write a function to reverse a linked list in - place??
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def reverse_linked_list(head):
    prev_node = None
    current_node = head
    
    while current_node is not None:
        next_node = current_node.next  # Store the next node
        current_node.next = prev_node  # Reverse the pointer
        
        # Move pointers one position ahead
        prev_node = current_node
        current_node = next_node
        
    # After the loop, prev_node will be the new head
    head = prev_node
    return head

# Example usage:
# Create a sample linked list: 1 -> 2 -> 3 -> 4 -> None
head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(4)

# Print original linked list
print("Original Linked List:")
current_node = head
while current_node is not None:
    print(current_node.data, end=" ")
    current_node = current_node.next
print()

# Reverse the linked list
head = reverse_linked_list(head)

# Print reversed linked list
print("Reversed Linked List:")
current_node = head
while current_node is not None:
    print(current_node.data, end=" ")
    current_node = current_node.next
```

```python
3) Detect the cycle in linked list??
```

```python
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None

def has_cycle(head):
    if not head or not head.next:
        return False

    slow = head
    fast = head.next

    while fast and fast.next:
        if slow == fast:
            return True
        slow = slow.next
        fast = fast.next.next

    return False

# Example usage:
# Create a sample linked list with a cycle: 1 -> 2 -> 3 -> 4 -> 5 -> 2
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(4)
head.next.next.next.next = ListNode(5)
head.next.next.next.next.next = head.next  # Pointing back to the second node

# Check if the linked list has a cycle
print(has_cycle(head))  # Output: True
```

```python
4) Merge two sorted linked list into one

1->3->5->6->null and 2->4->6->8->null should be merged to make

1->2->3->4->5->6->7->8
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def mergeTwoLists(l1, l2):
    dummy = ListNode(-1)
    current = dummy
    
    while l1 and l2:
        if l1.val < l2.val:
            current.next = l1
            l1 = l1.next
        else:
            current.next = l2
            l2 = l2.next
        current = current.next
    
    # Append remaining elements if any
    if l1:
        current.next = l1
    elif l2:
        current.next = l2
    
    return dummy.next

# Function to print the linked list
def printList(node):
    while node:
        print(node.val, end="->")
        node = node.next
    print("null")

# Test the function
if __name__ == "__main__":
    # Create the first linked list: 1->3->5->6->null
    l1 = ListNode(1)
    l1.next = ListNode(3)
    l1.next.next = ListNode(5)
    l1.next.next.next = ListNode(6)

    # Create the second linked list: 2->4->6->8->null
    l2 = ListNode(2)
    l2.next = ListNode(4)
    l2.next.next = ListNode(6)
    l2.next.next.next = ListNode(8)

    # Merge the two lists
    merged_list = mergeTwoLists(l1, l2)

    # Print the merged list
    print("Merged Linked List:")
    printList(merged_list)
```

```python
5) Write a function to remove nth node from the end in a linked list??

1->2->3->4->5->6, removing 2nd node from end will return 1->2->3->4->6.
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def removeNthFromEnd(head, n):
    dummy = ListNode(0)
    dummy.next = head
    first = dummy
    second = dummy
    
    # Move first pointer n steps ahead
    for _ in range(n + 1):
        first = first.next
    
    # Move both pointers until first reaches the end
    while first is not None:
        first = first.next
        second = second.next
    
    # Remove the nth node from the end
    second.next = second.next.next
    
    return dummy.next

# Function to print the linked list
def printList(node):
    while node:
        print(node.val, end="->")
        node = node.next
    print("null")

# Test the function
if __name__ == "__main__":
    # Create the linked list: 1->2->3->4->5->6
    head = ListNode(1)
    head.next = ListNode(2)
    head.next.next = ListNode(3)
    head.next.next.next = ListNode(4)
    head.next.next.next.next = ListNode(5)
    head.next.next.next.next.next = ListNode(6)

    # Remove the 2nd node from the end
    head = removeNthFromEnd(head, 2)

    # Print the modified linked list
    print("Modified Linked List:")
    printList(head)
```

```python
6)Remove duplicates from a sorted linked list??

1->2->3->3->4->4->4->5  should be changed to 1->2->3->4->5.
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def deleteDuplicates(head):
    current = head
    while current and current.next:
        if current.val == current.next.val:
            current.next = current.next.next
        else:
            current = current.next
    return head

# Function to print the linked list
def printList(node):
    while node:
        print(node.val, end="->")
        node = node.next
    print("null")

# Test the function
if __name__ == "__main__":
    # Create the linked list: 1->2->3->3->4->4->4->5
    head = ListNode(1)
    head.next = ListNode(2)
    head.next.next = ListNode(3)
    head.next.next.next = ListNode(3)
    head.next.next.next.next = ListNode(4)
    head.next.next.next.next.next = ListNode(4)
    head.next.next.next.next.next.next = ListNode(4)
    head.next.next.next.next.next.next.next = ListNode(5)

    # Remove duplicates
    head = deleteDuplicates(head)

    # Print the modified linked list
    print("Modified Linked List:")
    printList(head)
```

```python
7) Find the intersection of the two linked lists??

1->2->3->4->8->6->9  5->1->6->7  , intersection 1->6.
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def getIntersectionNode(headA, headB):
    if not headA or not headB:
        return None
    
    pointerA = headA
    pointerB = headB
    
    # Traverse both lists until they meet at the intersection or end of the lists
    while pointerA != pointerB:
        pointerA = pointerA.next if pointerA else headB
        pointerB = pointerB.next if pointerB else headA
    
    return pointerA

# Function to print the linked list
def printList(node):
    while node:
        print(node.val, end="->")
        node = node.next
    print("null")

# Test the function
if __name__ == "__main__":
    # Create the first linked list: 1->2->3->4->8->6->9
    headA = ListNode(1)
    headA.next = ListNode(2)
    headA.next.next = ListNode(3)
    headA.next.next.next = ListNode(4)
    headA.next.next.next.next = ListNode(8)
    headA.next.next.next.next.next = ListNode(6)
    headA.next.next.next.next.next.next = ListNode(9)

    # Create the second linked list: 5->1->6->7
    headB = ListNode(5)
    headB.next = headA.next.next  # intersect at node with value 3
    headB.next.next = headA.next.next.next.next  # intersect at node with value 8

    # Find the intersection node
    intersection = getIntersectionNode(headA, headB)

    # Print the intersection node
    if intersection:
        print("Intersection Node Value:", intersection.val)
    else:
        print("No Intersection")
```

```python
8) Rotate a linked list by k positions to the right??

 1->2->3->4->8->6->9 , after rotating for 2 times Cecomes , 3->4->8->6->9->1->2.
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def rotateRight(head, k):
    if not head or k == 0:
        return head
    
    # Find the length of the linked list
    length = 1
    tail = head
    while tail.next:
        length += 1
        tail = tail.next
    
    # Adjust k if it exceeds the length of the linked list
    k = k % length
    
    # No rotation needed if k is 0
    if k == 0:
        return head
    
    # Find the (length - k)th node
    new_tail_index = length - k - 1
    new_tail = head
    for _ in range(new_tail_index):
        new_tail = new_tail.next
    
    # Adjust pointers for rotation
    new_head = new_tail.next
    new_tail.next = None
    tail.next = head
    
    return new_head

# Function to print the linked list
def printList(node):
    while node:
        print(node.val, end="->")
        node = node.next
    print("null")

# Test the function
if __name__ == "__main__":
    # Create the linked list: 1->2->3->4->8->6->9
    head = ListNode(1)
    head.next = ListNode(2)
    head.next.next = ListNode(3)
    head.next.next.next = ListNode(4)
    head.next.next.next.next = ListNode(8)
    head.next.next.next.next.next = ListNode(6)
    head.next.next.next.next.next.next = ListNode(9)

    # Rotate the linked list by 2 positions to the right
    rotated_head = rotateRight(head, 2)

    # Print the rotated linked list
    print("Rotated Linked List:")
    printList(rotated_head)
```

```python
9) Add Two Numbers Represented by Linked Lists:
    
Given two non-empty linked lists representing two non-negative integers, where the digits are stored in 
reverse order, add the two numCers and return it as a linked list.
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def addTwoNumbers(l1, l2):
    dummy = ListNode()
    current = dummy
    carry = 0
    
    while l1 or l2 or carry:
        # Calculate sum and carry
        val1 = l1.val if l1 else 0
        val2 = l2.val if l2 else 0
        total = val1 + val2 + carry
        carry = total // 10
        current.next = ListNode(total % 10)
        
        # Move to next nodes
        current = current.next
        if l1:
            l1 = l1.next
        if l2:
            l2 = l2.next
    
    return dummy.next

# Function to print the linked list
def printList(node):
    while node:
        print(node.val, end="->")
        node = node.next
    print("null")

# Test the function
if __name__ == "__main__":
    # Create the first linked list: 2->4->3 (represents 342)
    l1 = ListNode(2)
    l1.next = ListNode(4)
    l1.next.next = ListNode(3)

    # Create the second linked list: 5->6->4 (represents 465)
    l2 = ListNode(5)
    l2.next = ListNode(6)
    l2.next.next = ListNode(4)

    # Add the two numbers represented by the linked lists
    result = addTwoNumbers(l1, l2)

    # Print the result as a linked list
    print("Sum Linked List:")
    printList(result)
```

```python
10) Clone a Linked List with next and Random Pointer :
    
Given a linked list of size N where each node has two links: one pointer points to the next node and the 
second pointer points to any node in the list. The task is to create a clone of this linked list in O(N) time. 


Note: The pointer pointing to the next node is ‘next‘ pointer and the one pointing to an arCitrary node is 
called ‘arCit’ pointer as it can point to any arCitrary node in the linked list.
```

```python
class Node:
    def __init__(self, val=0, next=None, random=None):
        self.val = val
        self.next = next
        self.random = random

def cloneLinkedList(head):
    if not head:
        return None
    
    # Create a hashmap to store mapping between original nodes and their clones
    node_map = {}
    
    # First pass: create clones and map original nodes to their clones
    current = head
    while current:
        node_map[current] = Node(current.val)
        current = current.next
    
    # Second pass: update next and random pointers of the clone nodes
    current = head
    while current:
        clone_node = node_map[current]
        clone_node.next = node_map.get(current.next)
        clone_node.random = node_map.get(current.random)
        current = current.next
    
    # Return the head of the cloned linked list
    return node_map[head]

# Function to print the linked list
def printLinkedList(head):
    visited = set()
    while head:
        if head in visited:
            print("Cycle detected!")
            break
        print("Value:", head.val, "Next:", head.next.val if head.next else "None", "Random:", head.random.val if head.random else "None")
        visited.add(head)
        head = head.next

# Test the function
if __name__ == "__main__":
    # Create the original linked list
    head = Node(1)
    head.next = Node(2)
    head.next.next = Node(3)
    head.next.next.next = Node(4)
    head.next.next.next.next = Node(5)

    # Assign random pointers
    head.random = head.next.next  # 1 -> 3
    head.next.random = head  # 2 -> 1
    head.next.next.random = head.next.next.next.next  # 3 -> 5
    head.next.next.next.random = head.next  # 4 -> 2
    head.next.next.next.next.random = head.next.next  # 5 -> 3

    # Clone the linked list
    cloned_head = cloneLinkedList(head)

    # Print the original and cloned linked lists
    print("Original Linked List:")
    printLinkedList(head)
    print("\nCloned Linked List:")
    printLinkedList(cloned_head)
```
